```python
def summary(df):
    analysis_data = []
    for col in df.columns:
        dtype = df[col].dtype
        missing_count = df[col].isnull().sum()
        missing_percentage = missing_count / len(df) * 100
        unique_count = df[col].nunique()
        top_7_values = df[col].value_counts().nlargest(7).index.tolist()
        analysis_data.append([col, dtype, missing_count, missing_percentage, unique_count, top_7_values])
    analysis_df = pd.DataFrame(analysis_data, columns=['Column', 'DataType', 'MissingCount', 'missing_percentage', 'UniqueCount', 'Top7Values'])
    return analysis_df
```

```python
import pandas as pd

output_drive_path = '/content/drive/MyDrive/01_남양주미래기술학교/자료'
df = pd.read_csv(output_drive_path+"/processed_adventureworks_sales_data.csv")
df
```

```python
df.columns
```

```python
ordered_columns = [

    # 날짜 정보
    'Date',
    #'Full Date',

    # 고객 및 리셀러 정보
    'Customer', 'Reseller', 'Business Type', 'Group',

    # 제품 정보
    'SKU', 'Product', 'Color', 'Model', 'Subcategory', 'Category',
    'Standard Cost', 'Product Standard Cost', 'List Price', 'Unit Price', 'Total Product Cost',

    # 주문 정보
    'Order Quantity', 'Sales Amount', 'Sales Order', 'Sales Order Line', 'Channel',

    # 지역 정보
    'City', 'State-Province', 'Country-Region', 'Country', 'Region', 'Postal Code',

]

df = df[ordered_columns]
```

```python
summary(df)
```

```python
df.loc[:, 'Date'] = pd.to_datetime(df['Date'])  # datetime 형 변환 안전하게

# Select columns with 'object' dtype
category_columns = df.select_dtypes(include='object').columns
for col in category_columns:
    df.loc[:, col] = df[col].astype('category')  # 범주형 안전 변환
```

```python
summary(df)
```

```python
# prompt: Customer를 총 기간 동안 Recency, Frequency, Monetary 기준으로 분석해줘, 주석을 한글로

# 필요한 라이브러리 임포트

import datetime as dt

# Recency, Frequency, Monetary 분석 함수 정의
# df: 분석할 데이터프레임 (반드시 'Customer', 'Date', 'Sales Amount' 컬럼 포함)
def rfm_analysis(df):
  # 분석 종료 날짜 설정 (데이터의 가장 최근 날짜 이후 하루)
  snapshot_date = df['Date'].max() + dt.timedelta(days=1)

  # 고객별 Recency, Frequency, Monetary 계산
  # Recency: snapshot_date 기준 가장 최근 구매일로부터의 경과 일수
  # Frequency: 총 구매 횟수
  # Monetary: 총 구매 금액
  rfm_df = df.groupby('Customer').agg({
      'Date': lambda date: (snapshot_date - date.max()).days, # Recency 계산
      'Sales Order': 'nunique',  # Frequency 계산 (고유한 주문 번호 수)
      'Sales Amount': 'sum'   # Monetary 계산 (총 판매 금액 합계)
  })

  # 컬럼 이름 변경
  rfm_df.rename(columns={'Date': 'Recency',
                         'Sales Order': 'Frequency',
                         'Sales Amount': 'Monetary'}, inplace=True)

  return rfm_df

# RFM 분석 실행
rfm_results = rfm_analysis(df)

# 결과 출력
print("RFM 분석 결과:")
rfm_results
```

```python

# Recency 값이 클수록 (더 이전에 구매) 낮은 점수를 부여하기 위해 내림차순 정렬 후 순위 할당
rfm_df['R_Score'] = pd.qcut(rfm_df['Recency'], 5, labels=[5, 4, 3, 2, 1])

# Frequency와 Monetary 값이 클수록 높은 점수를 부여하기 위해 오름차순 정렬 후 순위 할당
rfm_df['F_Score'] = pd.qcut(rfm_df['Frequency'], 5, labels=[1, 2, 3, 4, 5])
rfm_df['M_Score'] = pd.qcut(rfm_df['Monetary'], 5, labels=[1, 2, 3, 4, 5])

# RFM 점수 조합 (예: R_Score와 F_Score를 연결)
# M_Score도 포함하여 세 자리 점수 생성
rfm_df['RFM_Score'] = rfm_df['R_Score'].astype(str) + rfm_df['F_Score'].astype(str) + rfm_df['M_Score'].astype(str)

# 결과 출력
print(rfm_df.head())
print(rfm_df.describe())

# RFM 점수별 고객 세그먼트 정의 예시 (간단한 예시)
# 예: 555는 최고의 고객 (Champions), 111은 이탈 가능성이 높은 고객 (Lost)
# 실제 세그멘테이션은 비즈니스 목표에 따라 다르게 정의될 수 있습니다.

# RFM 세그먼트 예시 (추가적인 분석 및 시각화를 위해 세그먼트 컬럼 생성)
# 예시로 R과 F 점수를 조합하여 세그먼트 생성 (M도 고려 가능)
def segment_me(rfm):
    if rfm['R_Score'] == 5 and rfm['F_Score'] == 5:
        return 'Champions'
    elif rfm['R_Score'] == 5 and rfm['F_Score'] == 4:
        return 'Loyal Customers'
    elif rfm['R_Score'] == 4 and rfm['F_Score'] == 5:
         return 'Potential Loyalists'
    elif rfm['R_Score'] == 5 and rfm['F_Score'] == 1:
        return 'New Customers'
    elif rfm['R_Score'] == 4 and rfm['F_Score'] == 4:
        return 'Promising'
    elif rfm['R_Score'] == 3 and rfm['F_Score'] == 3:
        return 'Need Attention'
    elif rfm['R_Score'] == 2 and rfm['F_Score'] == 2:
        return 'At Risk'
    elif rfm['R_Score'] == 1 and rfm['F_Score'] == 1:
        return 'Lost'
    else:
        return 'Others'

rfm_df['Segment'] = rfm_df.apply(segment_me, axis=1)

# 세그먼트별 고객 수 확인
print("\n고객 세그먼트별 분포:")
print(rfm_df['Segment'].value_counts())

# 세그먼트별 평균 RFM 값 확인
print("\n세그먼트별 평균 RFM 지표:")
print(rfm_df.groupby('Segment')[['Recency', 'Frequency', 'Monetary']].mean())

# 결과 데이터프레임 확인
rfm_df.head()
```
