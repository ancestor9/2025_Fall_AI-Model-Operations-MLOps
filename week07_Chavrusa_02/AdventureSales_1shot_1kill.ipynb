```python
# prompt: AdventureWorks Sales.xlsx는 7개의 시트(고객, 상품, 지역, 판매 등)를 모두 읽고 사전형태로 데이터를 저장해 다오
# To read multiple sheets from a single Excel file, we will use the pandas library.
# Below is an example of how you can read all sheets from an Excel file into a dictionary of DataFrames.

import pandas as pd

# Function to read all sheets from an Excel file
def read_excel_sheets(excel_file):
    # Using sheet_name=None reads all sheets, each sheet as a DataFrame in a dictionary
    sheets_dict = pd.read_excel(excel_file, sheet_name=None)
    return sheets_dict

sheets_dict = read_excel_sheets('/content/AdventureWorks Sales.xlsx')
sheets_dict.keys()
```

```python
# @title
# prompt: https://github.com/microsoft/powerbi-desktop-samples/blob/main/AdventureWorks%20Sales%20Sample/AdventureWorks%20Sales.xlsx 파일은 여러 시트로 구성되어 있는데 모든 시트를 읽어줘

import pandas as pd
from tqdm import tqdm
#!pip install pandas openpyxl


# Replace with the actual URL of your Excel file
url = "https://github.com/microsoft/powerbi-desktop-samples/raw/main/AdventureWorks%20Sales%20Sample/AdventureWorks%20Sales.xlsx"

try:
  xls = pd.ExcelFile(url)
  sheet_names = xls.sheet_names

  all_sheets_data = {}
  for sheet_name in tqdm(sheet_names):
    df = pd.read_excel(url, sheet_name=sheet_name)
    print(df.head())
    all_sheets_data[sheet_name] = df
    print(f"Sheet '{sheet_name}' loaded successfully.")

except Exception as e:
  print(f"Error loading Excel file: {e}")
```

## **2. 데이터베이스에 저장하기**

```python
# prompt: all_sheets_data의 모든 데이터프레임을 sqlite3를 사용하여 DB를 만들고 각 테이블로 저장하라

import sqlite3

# Create a database connection
conn = sqlite3.connect('adventure_sales.db')

# Iterate through the dictionary of DataFrames and save each to a table
for sheet_name, df in all_sheets_data.items():
  # Replace spaces in sheet names with underscores for valid table names
  table_name = sheet_name.replace(" ", "_")
  df.to_sql(table_name, conn, index=False, if_exists='replace')
  print(f"DataFrame for sheet '{sheet_name}' saved to table '{table_name}'.")

# Close the connection
conn.close()
```

```python
# prompt: with 구문으로 코드를 만들어주라

# Create a database connection using the with statement
with sqlite3.connect('adventure_sales.db') as conn:
    # Iterate through the dictionary of DataFrames and save each to a table
    for sheet_name, df in all_sheets_data.items():
      # Replace spaces in sheet names with underscores for valid table names
      table_name = sheet_name.replace(" ", "_")
      df.to_sql(table_name, conn, index=False, if_exists='replace')
      print(f"DataFrame for sheet '{sheet_name}' saved to table '{table_name}'.")
```

## **EDA, 시각화, RFM 분석**

```python
# Merge Sales_Order_data and Customer_data DataFrames
# Let's check the columns of both dataframes to find a common key.
# print(all_sheets_data['Sales Order_data'].columns)
# print(all_sheets_data['Customer_data'].columns)

# Based on the column names, 'SalesOrderLineKey' seems to be the common key.
merged_df = pd.merge(all_sheets_data['Sales Order_data'], all_sheets_data['Sales_data'], on='SalesOrderLineKey', how='left')
merged_df = pd.merge(merged_df, all_sheets_data['Customer_data'], on='CustomerKey', how='left')

# Display the first few rows of the merged DataFrame
display(merged_df.head())
```

```python
# Remove rows with '[Not Applicable]' in any column
merged_df_cleaned = merged_df[~(merged_df == '[Not Applicable]').any(axis=1)]

# Display the number of rows before and after cleaning
print(f"Original number of rows: {len(merged_df)}")
print(f"Number of rows after removing '[Not Applicable]': {len(merged_df_cleaned)}")

# Display the first few rows of the cleaned DataFrame
display(merged_df_cleaned.head())
```

```python
merged_df_cleaned.columns
```

```python
# Group by 'Customer' and sum 'Sales Amount', then sort in descending order
customer_sales = merged_df_cleaned.groupby('Customer')['Sales Amount'].sum().sort_values(ascending=False)

# Display the result
display(customer_sales)
```

```python
vip = customer_sales[:100]
vip
```

```python
# Install gradio
# !pip install gradio

import gradio as gr

# Assuming 'vip' is already defined and is a pandas Series

def display_vip_customers(top_n):
  """Displays the top N VIP customers and their sales amount."""
  if top_n > len(vip) or top_n <= 0:
    return "Invalid number of customers."
  # Convert the Series to a DataFrame for better HTML representation in Gradio
  return vip.head(top_n).to_frame(name='Total Sales Amount').to_html()

# Create a Gradio interface
iface = gr.Interface(
    fn=display_vip_customers,
    inputs=gr.Slider(1, 100, value=10, step=1, label="Number of Top Customers"),
    outputs="html",
    title="Top VIP Customers by Sales Amount",
    description="Displays the top customers based on their total sales amount."
)

# Launch the Gradio interface
# Added share=True for Colab environment
iface.launch(inline=True, share=True)
```
